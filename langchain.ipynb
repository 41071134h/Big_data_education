```python
from langchain.llms import OpenAI
from langchain.chains.question_answering import load_qa_chain
from langchain.prompts import PromptTemplate
import pandas as pd

# 初始化 LLM（假設你有一個 OpenAI API key）
llm = OpenAI(openai_api_key="OpenAI_API_key")

# 讀取 CSV 檔案到 DataFrame
df = pd.read_csv("data.csv")

# 定義一個查詢模板，包括 context、df 和 question 三個變量
template = PromptTemplate(input_variables=["context", "df", "question"], template="{context}\n{df}\n{question}")

# 加載問答鏈
qa_chain = load_qa_chain(llm=llm, prompt=template)

# 定義查詢問題
question1 = "請問平均的 Quantity 是多少？"
question2 = "哪個 Method 的使用量最多？"

# 準備 context 和 DataFrame 作為字符串
context = "分析包含 Quantity 和 Method 的數據集"
df_string = df.to_string(index=False)

# 執行查詢，確保將 context、question 和 df 提供給 run() 函數
result1 = qa_chain.run(context=context, df=df_string, question=question1)
result2 = qa_chain.run(context=context, df=df_string, question=question2)

print("平均的 quantity 是:", result1)
print("使用量最多的 method 是:", result2)
```
